```python
df["genres"] = df["genres"].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
def remove_list(lst):
    if len(lst) == 0:
        return None
    else:
        return " ".join(lst)
```

```python
df = df.dropna(subset=['genres', 'type' , 'rating' , "synopsis" , "score" , "year"]).reset_index(drop=True)
```

```python
df["rating"] = df["rating"].replace(["PG-13 - Teens 13 or olde" , "G - All Ages" , "R - 17+ (violence & profanity)" , "Rx - Hentai" , "R+ - Mild Nudity" , "PG - Children"] , ["PG13" , "G" , "R" , "Rx" , "R+" , "PG"])
```

```python
df.head()
```

```python
scaler = MinMaxScaler()
num_vec =  scaler.fit_transform(df[['score', 'favorites', 'year']])
```

```python
num_vec
```

```python
df.head()
```

```python
com_text = df["type"] * 2 + ' ' + df["rating"] * 2 + ' '  + df["genres"] * 5 + ' ' + df["synopsis"] 
```

```python
com_text[0]
```

```python
com_text = com_text.str.replace("\n" , "")
com_text = com_text.str.replace("[Written by MAL Rewrite]" , "")
com_text = com_text.str.replace("Sci-Fi" , "SciFi")
```

```python
com_text = com_text.str.lower()
```

```python
com_text[0]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
com_text = com_text.apply(stem)
```

```python
com_text[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
cv = TfidfVectorizer(stop_words='english', max_features=5000)
text_vector = cv.fit_transform(com_text)
text_vector = text_vector.astype('float32')
```

```python
text_vector[0].toarray().shape
```

```python
from scipy.sparse import hstack
final_vector = hstack([text_vector, num_vec])
```

```python
final_vector.toarray()[0]
```

```python
list(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import linear_kernel
```

```python
similarity = linear_kernel(final_vector)
```

```python
data = pd.read_csv("cleaned_data.csv")
```

```python
def recommender(name):
    index = data[data["title"] == name].index[0]
    dis = similarity[index]
    movie_list = sorted(list(enumerate(dis)) , reverse = True , key = lambda x : x[1])[1:30]
    get_data = []
    for i in movie_list:
        temp = []
        temp.append(data.iloc[i[0]].id)
        temp.append(data.iloc[i[0]].title)
        temp.append(data.iloc[i[0]].Poster)
        temp.append(data.iloc[i[0]].score)

        get_data.append(temp)
    return get_data
```

```python
recommender("Naruto")
```

```python
import pickle
pickle.dump(similarity , open("similarity.pkl" , "wb"))
```
